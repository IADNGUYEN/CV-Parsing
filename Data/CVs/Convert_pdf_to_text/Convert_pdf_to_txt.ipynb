```python
#Using pymupdf
import fitz
import os
import re

files = []
path = '/content/drive/MyDrive/Nguyen Dinh Phuc Dai/Data/CVs/CVs_pdf/'
for f in os.listdir(path):
    if f.endswith('.pdf'):
        files.append(f)
        
path_output = '/content/drive/MyDrive/Nguyen Dinh Phuc Dai/Data/CVs/CVs_txt/'
for i in files:
    name = i.replace('.pdf', '')
    with open(path_output + name + '.txt','w',encoding = 'utf-8') as f:
        with fitz.open(path + i) as doc:
            text = ""
            for page in doc:
                text += page.get_text()
                text1 = re.sub(' *| *| *|• *| *| *| *|ü *|● *|� *| *| *|▪ *|· |➢||\s+-', ' ', text)
                resume = re.sub('\n\s+', '\n', text1)
            print(i)
            f.write(resume)
```
